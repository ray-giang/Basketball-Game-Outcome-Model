```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import string
import re
import numpy as np
from basketball_reference_web_scraper import client
from basketball_reference_web_scraper.data import OutputType
import datetime as dt
from datetime import datetime
```

```python
# web scraping 21 seasons of games, will remove playoff games in data manipulation portion of capstone project
all_season_results = pd.DataFrame()
for year in range(1999,2021):
    season_results = client.season_schedule(season_end_year=year,output_type=OutputType.CSV, output_file_path="{}_season_results.csv".format(year))
    season_results = pd.read_csv('{}_season_results.csv'.format(year))
    season_results['season'] = year
    all_season_results = all_season_results.append(season_results)
print(len(all_season_results))
```

```python
all_season_results2 = all_season_results.rename(columns={'start_time':'date',
                                                         'home_team':'team1',
                                                         'home_team_score':'score1',
                                                         'away_team':'team2',
                                                         'away_team_score':'score2',
                                                         'season':'season'})
all_season_results2['win_flag'] = np.where((all_season_results2['score1'] > all_season_results2['score2']),1,0)
all_season_results2['home_flag'] = 1
all_season_results2.head()
```

```python
# appending datasets together to have scheduled results based on a perspective of one team

all_season_results_final = all_season_results2.append(all_season_results1, sort=True)
print(len(all_season_results_final))
all_season_results_final.head()
```

```python
all_season_results_final.to_csv('all_season_results_final.csv')
```

```python
all_boxscore_results = pd.DataFrame()
for year in range(2020,2021):
    for month in (10,11,12,1,2,3,4,5,6):
        for day in range(1,32):
            boxscore_results = client.team_box_scores(year=year,month=month,day=day)
            boxscore_results = pd.DataFrame(boxscore_results)
            if len(boxscore_results)>0:
                boxscore_results['date'] = pd.to_datetime('{}-{}-{}'.format(year,month,day), format='%Y-%m-%d')          
                all_boxscore_results = all_boxscore_results.append(boxscore_results)
        print('{}-{}-{} has been created. There are now {} many records'.format(year,month,day,len(all_boxscore_results)))
    all_boxscore_results.to_csv('all boxscore results test {}.csv'.format(year))
print('Done, there are {} many records in the boxscore table)'.format(len(all_boxscore_results)))
all_boxscore_results.to_csv('all boxscore results test.csv')
```
